```python
df['Brand'] = df['Brand'].astype('category')
df['Model'] = df['Model'].astype('category')
df['Memory'] = df['Memory'].astype('category')
df['Storage'] = df['Storage'].astype('category')  
  
# Assigning numerical values and storing it in another columns
df['Gen_new'] = df['Brand'].cat.codes
df['Rem_new'] = df['Model'].cat.codes
df['rew_new'] = df['Memory'].cat.codes
df['nem_new'] = df['Storage'].cat.codes  
  
# Create an instance of One-hot-encoder
enc = OneHotEncoder()
  
# Passing encoded columns
  
enc_data = pd.DataFrame(enc.fit_transform(
    df[['Gen_new', 'Rem_new','rew_new','nem_new']]).toarray())
  
# Merge with main
New_df = df.join(enc_data)
  
print(New_df)
```
